Data Preparation

```python
movies_release.duplicated().sum()
```

```python
movies_release.drop_duplicates(inplace=True)
```

```python
movies_release.duplicated().sum()
```

```python
movies_release.isna().sum()
```

```python
movies_release["belongs_to_collection"]=movies_release["belongs_to_collection"].fillna("No Data")
movies_release["homepage"]=movies_release["homepage"].fillna("No Data")
movies_release["tagline"]=movies_release["tagline"].fillna("No tagline")
```

```python
movies_release.dropna(inplace=True)
```

```python
movies_release.isna().sum()
```

```python
os.listdir()
```

```python
rating=pd.read_csv("ratings_small.csv")
```

```python
rating
```

```python
rating.info()
```

```python
rating["timestamp"]=pd.to_datetime(rating["timestamp"], unit="s")
```

```python
rating["timestamp"]
```

```python
rating_movies=rating.groupby("movieId", as_index=False)["rating"].median()
```

```python
rating_movies
```

Penambahan atribut "rating" di dapatkan dari dataset yang berbeda. serta penggabungan sesuai atribut "movieId" dari dataset asli

```python
movies_release=movies_release.merge(rating_movies,
                     left_on="id",
                     right_on="movieId",
                     how="inner")
```

```python
movies_release=movies_release.sort_values(["budget", "runtime", "revenue", "vote_average", "rating"], ascending=[False, False, False, False, False])
```

```python
movies_release
```

Memberikan atribut tambahan "profit" agar mengetahui berapa keuntungan hasil dari pembuatan film

```python
movies_release["profit_movies"]=movies_release["revenue"].sub(movies_release["budget"])
```

```python
movies_release[["original_title", "profit_movies", "revenue", "runtime", "vote_average", "rating"]].sort_values("profit_movies", ascending=False)
```

Informasi Film Durasi Lebih dari 3 Jam (Feature Engineering)

```python
movies_release["movies_hight_duration"]=(movies_release["runtime"]>= 60*3).astype("int")
```

```python
movies_release[movies_release["movies_hight_duration"]==1]
```

Informasi Judul Film Terpanjang - Terpendek (Feature Engineering)

```python
movies_release["movies_length"]=movies_release["original_title"].str.len()
```

```python
movies_release[movies_release["movies_length"]<5]
```

Informasi Jumlah Pembuatan Film Berdasarkan Tahun (Feature Engineering)

```python
movies_release["movies_years"]=movies_release["release_date"].dt.year
```

```python
years_movies=movies_release.groupby("movies_years").size()
```

```python
years_movies.sort_values(ascending=False)
```

```python
sns.scatterplot(data=movies_release,
             x="runtime",
             y="profit_movies")
plt.show()
```

Tahap EDA - Agar informasi profit film yang ditampilkan hanya yang mendapatkan keuntungan > 0

```python
movies_profit=movies_release[movies_release["profit_movies"]>0]
```

```python
sns.scatterplot(movies_profit,
                x="runtime",
                y="profit_movies")
plt.show()
```

Tahap EDA - Agar informasi rating film yang ditampilkan > 3

```python
movies_data_new=movies_profit.merge(rating_movies,
                    left_on="id",
                    right_on="movieId",
                    how="inner")
```
