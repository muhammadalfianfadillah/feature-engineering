```python
movies_data=pd.read_csv("movies_metadata.csv")
```

```python
movies_data["release_date"]=pd.to_datetime(movies_data["release_date"])
```

```python
movies_data["release_date"]
```

```python
movies_data=movies_data.set_index("original_title")
```

```python
movies_data=movies_data.reset_index("original_title")
```

```python
movies_release_en=movies_data[(movies_data["status"]=="Released")&(movies_data["original_language"]=="en")].sort_values(["budget", "revenue", "runtime", "vote_average"], ascending=[False, False, False, False])
```

```python
movies_release_en
```

```python
movies_release_en.duplicated().sum()
```

```python
movies_release_en.drop_duplicates(inplace=True)
```

```python
movies_release_en.duplicated().sum()
```

```python
movies_release_en.isna().sum()
```

```python
movies_release_en["belongs_to_collection"]=movies_release_en["belongs_to_collection"].fillna("No Data")
movies_release_en["homepage"]=movies_release_en["homepage"].fillna("No Data")
movies_release_en["tagline"]=movies_release_en["tagline"].fillna("No tagline")
```

```python
movies_release_en.isna().sum()
```

```python
movies_release_en.dropna(inplace=True)
```

```python
movies_release_en.isna().sum()
```

```python
os.listdir()
```

```python
rating=pd.read_csv("ratings_small.csv")
```

```python
rating
```

```python
rating.info()
```

```python
rating["timestamp"]=pd.to_datetime(rating["timestamp"], unit="s")
```

```python
rating["timestamp"]
```

```python
rating_movies=rating.groupby("movieId")["rating"].agg(["mean", "count", "median"])
```

```python
rating_movies
```

```python
movies_release_en=movies_release_en.merge(rating_movies,
                                          left_on="id",
                                          right_on="movieId",
                                          how="inner")
```

```python
movies_release_en
```

```python
movies_release_en["profit"]=movies_release_en["revenue"].sub(movies_release_en["budget"])
```

```python
movies_release_en[["original_title", "profit", "budget", "revenue", "runtime", "vote_average"]].sort_values("profit", ascending=False)
```

```python
movies_release_en["movies_duration_hight"]=(movies_release_en["runtime"]>=60*3).astype("int")
```

```python
movies_release_en[movies_release_en["movies_duration_hight"]==1].sort_values(["original_title", "profit", "revenue", "runtime", "vote_average"], ascending=[False, False, False, False, False])
```

```python
movies_release_en["movies_title_length"]=movies_release_en["original_title"].str.len()
```

```python
movies_release_en[movies_release_en["movies_title_length"]<5].sort_values(["original_title", "profit", "revenue", "runtime", "vote_average"], ascending=[False, False, False, False, False])
```

```python
movies_release_en["movies_favorit"]=movies_release_en["median"]/60
```

```python
movies_release_en["movies_years"]=movies_release_en["release_date"].dt.year
```

```python
movies_years=movies_release_en.groupby("movies_years").size()
```

```python
movies_years.sort_values(ascending=False)
```
